```python
import langchain
import sentence_transformers
import faiss
import gradio

print("LangChain:", langchain.__version__)
print("Sentence Transformers:", sentence_transformers.__version__)
print("FAISS Installed Successfully")
print("Gradio:", gradio.__version__)
```

```python
knowledge_text = """
Artificial Intelligence (AI) is the simulation of human intelligence by machines.

Machine Learning (ML) is a subset of AI that enables computers to learn from data without being explicitly programmed.

Deep Learning is a subset of Machine Learning that uses neural networks with multiple layers.

Natural Language Processing (NLP) enables computers to understand and generate human language.

Computer Vision enables machines to interpret and analyze images and videos.

Supervised Learning uses labeled data to train models.

Unsupervised Learning discovers hidden patterns in unlabeled data.

Reinforcement Learning trains agents using rewards and penalties.

BERT (Bidirectional Encoder Representations from Transformers) is a transformer-based language model developed by Google.

RAG (Retrieval-Augmented Generation) combines information retrieval with language generation to produce context-aware responses.

LangChain is a framework for developing applications powered by large language models.

FAISS is a library developed by Meta for efficient similarity search and vector retrieval.

Embeddings are numerical vector representations of text used to measure semantic similarity.

Vector Databases store embeddings and retrieve the most relevant information for user queries.
"""

print("Knowledge Base Created Successfully!")
```

```python
from langchain_core.documents import Document

# Convert text into LangChain Document
documents = [Document(page_content=knowledge_text)]

print("Documents Created Successfully!")
print(f"Number of Documents: {len(documents)}")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Create text splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=50
)

# Split documents into chunks
chunks = text_splitter.split_documents(documents)

print("Documents Split Successfully!")
print(f"Total Chunks: {len(chunks)}")
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

# Load embedding model
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Create FAISS vector database
vectorstore = FAISS.from_documents(
    documents=chunks,
    embedding=embeddings
)

print("FAISS Vector Database Created Successfully!")
```

```python
# Create Retriever

retriever = vectorstore.as_retriever(
    search_kwargs={"k": 3}
)

print("Retriever Created Successfully!")
```

```python
# Test the Retriever

query = "What is BERT?"

retrieved_docs = retriever.invoke(query)

print("Retrieved Documents:\n")

for i, doc in enumerate(retrieved_docs, 1):
    print(f"Document {i}:")
    print(doc.page_content)
    print("-" * 50)
```

```python
# Simple RAG Chatbot Function

def chatbot(question):
    docs = retriever.invoke(question)

    if not docs:
        return "Sorry, I couldn't find relevant information."

    answer = "\n\n".join([doc.page_content for doc in docs])

    return answer

print("Chatbot Function Created Successfully!")
```

```python
# Test the Chatbot

question = "What is Machine Learning?"

response = chatbot(question)

print("Question:")
print(question)

print("\nAnswer:")
print(response)
```

```python
import gradio as gr

demo = gr.Interface(
    fn=chatbot,
    inputs=gr.Textbox(
        lines=4,
        placeholder="Ask a question about AI or Machine Learning..."
    ),
    outputs="text",
    title="Context-Aware Chatbot using RAG",
    description="Ask questions related to Artificial Intelligence and Machine Learning."
)

demo.launch()
```
